```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("output_Fresh.tsv", delimiter = '\t', quoting = 3)
dataset.head()
```

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
p = PorterStemmer()

all_swords = stopwords.words('english')
all_swords.remove('not')
```

```python
corpus=[]



for i in range (0,100):
  review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  review = review.lower()
  review = review.split()
  review = [p.stem(word) for word in review if not word in set(all_swords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle
cvFile = 'Bag_of_Words_Sentiment_Model.pkl'
cv = pickle.load(open(cvFile, "rb"))
```

```python
X_fresh = cv.transform(corpus).toarray()
X_fresh.shape
```

```python
# predictions from sentiment classifier

import joblib
classifier = joblib.load('classifier')
```

```python
from sklearn.linear_model import LogisticRegression
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
dataset['predicted_label'] = y_pred.tolist()
dataset.head()
```

```python
dataset.to_csv("output_Fresh_Sentiments.tsv", sep='\t', encoding='UTF-8', index=False)
```
